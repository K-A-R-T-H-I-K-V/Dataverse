In Python programming, the functions make_circles and make_moons are part of scikit-learn, a popular machine learning library. These functions are used to create synthetic datasets that are often employed to test and demonstrate machine learning algorithms, especially classification and clustering.<br>'make_circles' produces Gaussian data with a spherical decision boundary for binary classification, while 'make_moons' produces two interleaving half circles.

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=750, factor=0.3, noise=0.1) 
# n_samples : total number of points generated
# factor : Scale factor between inner and outer circle in the range [0, 1)
# noise : noise is often added artificially to data to simulate real-world imperfections or to make models more robust
```

```python
X, y
```

```python
plt.scatter(X[:,0], X[:,1])

# Separates the dimensions (axes) of the array.
''' The part before the comma deals with rows (1st dimension), 
and the part after the comma deals with columns (2nd dimension) '''
```

After selecting a candidate eps, check:

<br>Number of Clusters:
<br>Too many clusters → eps might be too small.
<br>One big cluster → eps might be too large.
<br>Noise Points:
<br>If too many points are labeled as noise (-1), increase eps

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.15)
dbscan.fit_predict(X)

'''In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), the parameter eps (short for epsilon) is a key hyperparameter, 
that determines the maximum distance between two points for them to be considered as part of the same neighborhood.
'''
```

```python
dbscan.labels_
```

After running the fit method on your dataset, the labels_ attribute contains an array where each element corresponds to the cluster label of a data point.
Values in labels_:
<br>Non-negative integers (0, 1, 2, ...): Indicate the cluster to which a data point belongs.
<br>-1: Indicates a point classified as noise (not part of any cluster).

```python
plt.scatter(X[:,0], X[:,1], c=dbscan.labels_)
```
