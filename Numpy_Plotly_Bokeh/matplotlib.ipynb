```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
print(x)
```

```python
y = np.sin(x)
print(y)
```

```python
plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('F(x) = y = sin(x)')

plt.title('Sine')
plt.show()
```

```python
# plt.figure(figsize=(8, 4))
plt.plot(x, y, '--r')

plt.xlabel('x')
plt.ylabel('F(x) = y = sin(x)')

plt.grid()
plt.title('Sine')
plt.show()
```

```python
x1 = np.random.rand(50)
y1 = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x1, y1, c = colors, s = sizes, alpha = 0.4)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
x1 = np.random.rand(50)
y1 = np.random.rand(50)
# colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x1, y1, c = 'c', s = sizes, alpha = 0.8, marker='x')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
x2 = ['A', 'B', 'C', 'D', 'E']
y2 = np.random.rand(5)

plt.bar(x2, y2)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Plot')
plt.show()
```

```python
x2 = ['A', 'B', 'C', 'D', 'E']
y2 = np.random.rand(5)

plt.barh(x2, y2)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Plot')
plt.show()
```
