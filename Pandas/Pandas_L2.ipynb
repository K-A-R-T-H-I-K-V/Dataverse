## '<u>describe()</u>'
#### The describe() function in Pandas provides a quick summary of the statistical properties of the numerical columns in a DataFrame or Series. It generates descriptive statistics that give an overview of the central tendencies, dispersion, and shape of the dataset's distribution. This function is commonly used for exploratory data analysis to get a sense of the data's basic characteristics.

## What describe() Provides
#### Count: The number of non-null (non-missing) entries in each column.
#### Mean: The average value of the data in each column.
#### Standard Deviation (std): Measures the amount of variation or dispersion from the mean.
#### Min: The minimum value in each column.
#### 25th Percentile (25%): The value below which 25% of the data falls (first quartile).
#### 50th Percentile (50%): The median value, which separates the data into two equal halves.
#### 75th Percentile (75%): The value below which 75% of the data falls (third quartile).
#### Max: The maximum value in each column.

#### In Pandas, <u>Categorical</u> is a data type used to represent categorical data, which can be particularly useful for variables that take on a limited and fixed number of possible values. Categorical data often includes factors or labels like gender, geographic regions, or grades. Using the Categorical type can be beneficial for both memory efficiency and performance when working with categorical variables.

## Key Features of Categorical
#### Efficient Storage: Categorical is more memory-efficient compared to object types, especially when there are many repeated values. Instead of storing repeated values as strings, it stores them as integer codes along with a corresponding list of categories.
#### Ordered vs. Unordered: Categorical supports both ordered and unordered categories. You can specify whether the categories have a natural order (e.g., 'small', 'medium', 'large') or if they are just labels without any inherent order.
#### Enhanced Performance: Operations involving categorical data can be faster due to the internal integer representation.
#### Integration with Other Features: Categorical integrates well with Pandas functionalities like grouping, pivot tables, and plotting, making it suitable for various data analysis tasks.

#### In Pandas, the iloc indexer is used for integer-location based indexing and selection. It allows you to select rows and columns from a DataFrame or Series by their integer positions (i.e., by their index positions, not by labels). This is particularly useful when you want to access specific rows or columns by their position in the DataFrame or Series, regardless of their actual labels.

## Key Features of iloc
#### Integer-Based Indexing: iloc uses zero-based integer indexing to select rows and columns. For example, iloc[0] selects the first row, and iloc[:, 1] selects the second column.
#### Range-Based Selection: You can use slicing to select a range of rows or columns. For example, iloc[1:4] selects rows 1 through 3.
#### Mixed Indexing: You can use iloc to select specific rows and columns. For example, iloc[0:3, 1:4] selects rows 0 through 2 and columns 1 through 3

```python
# df.iloc[row_indexer, column_indexer]
# row_indexer: Integer or slice to select specific rows.
# column_indexer: Integer or slice to select specific columns
```

```python
df.iloc[0:2]
```

```python
df.iloc[0:2, ['PassengerId', 'Survived', 'Pclass']]
```

#### iloc is meant for integer-based indexing. This means it requires integer positions for both rows and columns.
#### The code ['PassengerId', 'Survived', 'Pclass'] is a list of column names, which is not valid for iloc. Instead, iloc expects integer positions

```python
df.loc[0:2, ['PassengerId', 'Survived', 'Pclass']]
```

```python
# iloc -> integer location, does't recognize column names
# Internally Python assigns indices for column names as well
df.iloc[0:2, [0, 1, 2]]
```

```python
df.iloc[0:2, 0:3]
```
